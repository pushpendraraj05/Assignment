```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
```

```python
movie=pd.read_csv("D:\\assignment\\association rule\\my_movies.csv")
movie
```

```python
movie=pd.get_dummies(movie)
movie
```

```python
movie.shape
```

## Apriori Algorithm

```python
movie_ap=apriori(movie,min_support=0.4,use_colnames=True)
movie_ap
```

```python
movie_ap.sort_values("support", ascending = True)
```

```python
rule =association_rules(movie_ap, metric ="confidence", min_threshold = 0.7)
rule
```

```python
rule.sort_values('support', ascending = False).head(15)
```

```python
rule.sort_values('confidence', ascending = False).head(15)
```
